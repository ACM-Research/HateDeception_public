```
hateAugmenter = EasyDataAugmenter(pct_words_to_swap=0.2, transformations_per_example=1)

hateSpeechDf = pd.read_csv('changedData_hate_tweets.csv')
hateSpeechDf.head()
```

```
augHateDf = pd.DataFrame(columns=['0'])
for eachTweet in hateSpeechDf.tweet:
    #print(eachTweet, "-----------", hateAugmenter.augment(eachTweet))
    augHateDf = augHateDf.append(hateAugmenter.augment(eachTweet))

augHateDf.to_csv('augmented_changedData.csv',index = False)
```
