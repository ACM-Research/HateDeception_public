```
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```
# read in tweets
df = pd.read_csv('labeled_data.csv')
tweets = df['tweet'].tolist()

# load pretrained glove embeddings file
glove_file = 'glove.6B.50d.txt'
word_embeddings = {}

with open(glove_file, 'r') as f:
    for line in f:
        values = line.split()
        word = values[0]
        embedding = np.asarray(values[1:], dtype='float32')
        word_embeddings[word] = embedding
```

```
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
```

```
# preprocessing the tweets
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
preprocessed_tweets = []

# remove stopwords and extra characters
# tokenize tweet
for tweet in tweets:
    tweet = re.sub(r'#\w+|\@\w+|http\S+', '', tweet)
    tokens = word_tokenize(tweet)
    tokens = [token for token in tokens if token.lower() not in stop_words]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    preprocessed_tweets.append(tokens)
```

```
modified_tweets = []
for tweet in preprocessed_tweets:
    modified_tokens = []
    for token in tweet:
        if token in word_embeddings:
            embedding = word_embeddings[token]
            sim_words = [word for word, sim in sorted(word_embeddings.items(),
                                                      key=lambda x: np.dot(x[1], embedding),
                                                      reverse=True)[:10]]
            pos = nltk.pos_tag([token])[0][1]
            for sim_word in sim_words:
                if nltk.pos_tag([sim_word])[0][1] == pos:
                    modified_tokens.append(sim_word)
                    break
        else:
            modified_tokens.append(token)
    modified_tweet = ' '.join(modified_tokens)
    modified_tweets.append(modified_tweet)
```

```
word_embedded_tweets = 'word_embedding.csv'
newdf = pd.DataFrame(modified_tweets)
newdf.to_csv(word_embedded_tweets, index=False)
```
